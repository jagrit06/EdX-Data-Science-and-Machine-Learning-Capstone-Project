Obtaining and Cleaning Data

```python
#Selecting appropriate fields from the pluto data
fields = ['Address', 'BldgArea', 'BldgDepth', 'BuiltFAR', 'CommFAR', 'FacilFAR', 'Lot', 'LotArea', 
          'LotDepth', 'NumBldgs', 'NumFloors', 'OfficeArea', 'ResArea', 'ResidFAR', 
          'RetailArea', 'YearBuilt', 'YearAlter1', 'ZipCode', 'YCoord', 'XCoord']

df_BS = df_BS[fields]
df_BX = df_BX[fields]
df_MN = df_MN[fields]
df_QN = df_QN[fields]
df_SI = df_SI[fields]

#Creating one dataframe containing data on the buildings
df_bldg = pd.concat([df_BS, df_BX, df_MN, df_QN, df_SI])  
```

```python
df_bldg.dropna()  
df_bldg.reset_index(drop=True, inplace=True)  #Resetting indicies
df_bldg.head()
```

Since the XCoord, YCoord and ZipCode both depict the location, only the ZipCode will be taken in the model 

```python
df_bldg = df_bldg.drop(columns = ['XCoord', 'YCoord'])
```

```python
df_bldg.dtypes
```

```python
df_bldg.dropna(subset=['ZipCode'], axis=0, inplace=True)
df_bldg.reset_index(drop=True, inplace=True)

#Fixing data types
df_bldg['ZipCode'] = df_bldg['ZipCode'].astype(int).astype('category')
df_bldg['Lot'] = df_bldg['Lot'].astype('category')
df_bldg['LotArea'] = df_bldg['LotArea'].astype(float)
df_bldg['OfficeArea'] = df_bldg['OfficeArea'].astype(float)
df_bldg['ResArea'] = df_bldg['ResArea'].astype(float)
df_bldg['OfficeArea'] = df_bldg['OfficeArea'].astype(float)
df_bldg['BldgArea'] = df_bldg['BldgArea'].astype(float)
```

```python
df_311_2 = df_311.loc[:,['created_date', 'unique_key', 'complaint_type', 'incident_zip', 'incident_address', 'borough']] #taking dataframe with relevant details

#fixing datatype
df_311_2['created_date'] = pd.to_datetime(df_311['created_date'])
df_311_2['complaint_type'] = df_311['complaint_type'].astype('category')
df_311_2['incident_zip'] = df_311['incident_zip'].astype('category')
df_311_2['borough'] = df_311['borough'].astype('category')

#Taking complaints created before 31 Dec 2018
df_311_2 = df_311_2[df_311_2['created_date'] < pd.datetime(2019,1,1)]

#Taking top complaint as determined in Question 1
df_top_comp = pd.concat([df_311_2[df_311_2['complaint_type'] == 'HEATING'], df_311_2[df_311_2['complaint_type'] == 'HEAT/HOT WATER']])

#Dropping invalid data
df_top_comp.dropna()
df_top_comp.reset_index(drop=True, inplace=True)

df_top_comp.head()
```

```python
#Making a DataFrame of addresses that report the complaint of the type determined in Question 1
complaint_addresses = pd.DataFrame(df_top_comp['incident_address'].unique())

#Making Boolean array which checks if the address of the building in the building dataframe has registered a complaint of the type determined in Question 1
complaint = df_bldg['Address'].index.isin(complaint_addresses.index)

#Adding the boolean array as a column to building array. This is what needs to be studied and predicted 
df_bldg['Complaint'] = complaint
df_bldg = df_bldg.dropna()
df_bldg.reset_index(drop=True, inplace=True)

df_bldg.head(5)
```

Analysizing the Data to determine what building features relate to a complaint of the top complaint type being reported

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Using Perason correaltion to explore the data
plt.figure(figsize=(15,10))
sns.heatmap(df_bldg.corr(), annot = True, cmap=plt.cm.Reds)
plt.show()
```

```python
pear_coeff = abs(df_bldg.corr()['Complaint'])
pear_coeff.plot(kind = 'barh', figsize = (15,10))
plt.ylabel('Feature')
plt.xlabel('Absolute Value of Pearson Coefficient')
```

According to the pearson correlation, important features are the  ResidFAR, FacilFAR

It is important to that the pearson correlation does not comment on the ZIP code becuase the zip code is an arbitrarily assigned categorical value and hence should not be treated as a number on its own, and so other models must be employed as well so incorporate the ZIP Code

Since we wish to predict if a building, given its features, is likely to report a complaint of the type determined in question 1, we need to use an appropriately suited Classification Algorithim 

Since we expect categories to be in the features set and wish to explore the importance of the features used, decision trees are used (including a XGBoost model)

```python
y = df_bldg['Complaint']  #Making target array for predicitions
y[0:5]
```

```python
df_bldg = df_bldg.drop(columns = ['Address', 'Complaint'])

X = df_bldg.values #Making features array for predicitions
X[0:5]
```

```python
#Creating training and test data sets
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.35, random_state=3)
```

```python
#Making the decision tree
from sklearn.tree import DecisionTreeClassifier 
from sklearn import metrics
drugTree = DecisionTreeClassifier(criterion="entropy")
drugTree.fit(X_trainset,y_trainset)
predTree = drugTree.predict(X_testset)
print("Decision Trees's Accuracy: %.3f" %metrics.accuracy_score(y_testset, predTree))
```

```python
importances = drugTree.feature_importances_

# Ploting the feature importances of the decision tree
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances, color="r", align="center")
plt.xticks(range(X.shape[1]), df_bldg.columns, rotation='vertical')
plt.show()
```

```python
#Making XGBoost Model
import xgboost as xgb
model=xgb.XGBClassifier(random_state=1,learning_rate=0.05)
model.fit(X_trainset, y_trainset)
print('The accuracy of the XGBoost model is %.3f' %model.score(X_testset,y_testset))
```

```python
#Plotting feature importance from XGBoost Model
from xgboost import plot_importance
plot_importance(model)
plt.show()
```

```python
#Getting the indicies of the important features 
imp_features = [16,12]
print('According to the decision tree and XGBoost feature importance, important features are: \n')
for i in imp_features:
    print(df_bldg.columns[i] + '\n')
```
